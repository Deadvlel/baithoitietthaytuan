```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Tải dữ liệu từ tệp load.csv
# Đảm bảo tệp load.csv nằm trong cùng thư mục với notebook
try:
    df = pd.read_csv('load.csv')
except FileNotFoundError:
    print("Lỗi: Không tìm thấy tệp 'load.csv'. Vui lòng kiểm tra lại đường dẫn.")
    exit()

# Cấu hình các tên cột dựa trên dữ liệu bạn cung cấp
TIME_COLUMN = 'date'
variables = ['temperature', 'relative_humidity', 'precipitation']
```

```python
# Chuyển cột thời gian sang định dạng datetime và đặt làm Index
df[TIME_COLUMN] = pd.to_datetime(df[TIME_COLUMN])
df = df.set_index(TIME_COLUMN)

print("Thông tin DataFrame sau khi set Index:")
df.info()
```

```python
# Đặt style và figure size chung
sns.set_style("darkgrid")
plt.rcParams["figure.figsize"] = (12, 5)

# Hàm vẽ biểu đồ Line và Scatter
def plot_time_series(df, variable, freq, chart_type):
    # Resample và tính giá trị trung bình (mean) cho tuần/tháng
    if freq == 'W':
        resampled_df = df[variable].resample('W').mean().dropna()
        title_suffix = f"theo Tuần (Trung bình)"
    else: # freq == 'M'
        resampled_df = df[variable].resample('M').mean().dropna()
        title_suffix = f"theo Tháng (Trung bình)"

    if resampled_df.empty:
        print(f"Không đủ dữ liệu sau khi Resample {freq} cho {variable} để vẽ {chart_type}.")
        return

    plt.figure(figsize=(12, 5))
    
    if chart_type == 'line':
        plt.plot(resampled_df.index, resampled_df.values, label=variable, marker='o', linestyle='-')
        plt.title(f"a) Line Chart: {variable.replace('_', ' ').title()} {title_suffix}")
    elif chart_type == 'scatter':
        plt.scatter(resampled_df.index, resampled_df.values, label=variable, alpha=0.6)
        plt.title(f"b) Scatter Plot: {variable.replace('_', ' ').title()} {title_suffix}")

    plt.xlabel("Thời Gian")
    plt.ylabel(variable.replace('_', ' ').title())
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Hàm vẽ biểu đồ Histogram (Phân phối của giá trị trung bình theo Tuần/Tháng)
def plot_histogram(df, variable, freq):
    
    if freq == 'W':
        data = df[variable].resample('W').mean().dropna()
        title_suffix = "Tuần (Giá trị trung bình)"
    else: # freq == 'M'
        data = df[variable].resample('M').mean().dropna()
        title_suffix = "Tháng (Giá trị trung bình)"
        
    if data.empty:
        print(f"Không đủ dữ liệu sau khi Resample {freq} cho {variable} để vẽ Histogram.")
        return

    plt.figure(figsize=(10, 5))
    # Điều chỉnh bins cho phù hợp với số lượng điểm dữ liệu
    bins_count = min(15, len(data) // 2) if len(data) > 10 else 5 
    sns.histplot(data, kde=True, bins=bins_count)
    plt.title(f"c) Histogram: Phân phối {variable.replace('_', ' ').title()} theo {title_suffix}")
    plt.xlabel(variable.replace('_', ' ').title())
    plt.ylabel("Tần suất")
    plt.tight_layout()
    plt.show()


# --- Vòng lặp chính để tạo 18 biểu đồ ---
for var in variables:
    # Bỏ qua các cột không phải dạng số (đề phòng)
    if pd.api.types.is_numeric_dtype(df[var]): 
        print(f"\n==================== Biến: **{var.replace('_', ' ').title()}** ====================")
        
        # 1. LINE CHART (Tuần & Tháng)
        plot_time_series(df, var, 'W', 'line') 
        plot_time_series(df, var, 'M', 'line') 

        # 2. SCATTER PLOT (Tuần & Tháng)
        plot_time_series(df, var, 'W', 'scatter') 
        plot_time_series(df, var, 'M', 'scatter') 

        # 3. HISTOGRAM (Tuần & Tháng)
        plot_histogram(df, var, 'W')
        plot_histogram(df, var, 'M')
    else:
         print(f"Cột '{var}' không phải là dạng số và đã bị bỏ qua.")
```

```python
# Thiết lập các biến và cửa sổ trung bình trượt
variables = ['temperature', 'relative_humidity', 'precipitation']
ROLLING_WINDOW = 3 # Cửa sổ 3 tháng (có thể thay đổi)
FREQ = 'M' # Tần suất re-sample: 'M' cho Tháng, 'W' cho Tuần

# Thiết lập style và figure size
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (14, 6)

# --- Vòng lặp vẽ đồ thị xu hướng ---
for var in variables:
    # Kiểm tra đảm bảo cột là dạng số
    if pd.api.types.is_numeric_dtype(df[var]):
        print(f"\n==================== Xu hướng: **{var.replace('_', ' ').title()}** ====================")
        
        # 1. Resample dữ liệu về giá trị trung bình theo Tháng
        # Resample ('M') nếu dữ liệu đủ dày, nếu không thì dùng 'W' (Tuần)
        resampled_mean = df[var].resample(FREQ).mean().dropna()

        # Kiểm tra dữ liệu sau resample
        if len(resampled_mean) < ROLLING_WINDOW:
             # Nếu dữ liệu resample quá ít, không thể tính Rolling Mean.
             print(f"Dữ liệu sau resample ({FREQ}) quá thưa. Đang chuyển sang tần suất Tuần (W).")
             FREQ_ALT = 'W'
             resampled_mean = df[var].resample(FREQ_ALT).mean().dropna()
             # Đặt cửa sổ nhỏ hơn cho Tuần (ví dụ 4 tuần = 1 tháng)
             ROLLING_WINDOW_ALT = 4 
             rolling_mean = resampled_mean.rolling(window=ROLLING_WINDOW_ALT, center=True).mean()
             freq_text = f"Tuần + {ROLLING_WINDOW_ALT} Tuần Rolling Mean"
        else:
             # 2. Tính Rolling Mean (Trung bình trượt)
             rolling_mean = resampled_mean.rolling(window=ROLLING_WINDOW, center=True).mean()
             freq_text = f"Tháng + {ROLLING_WINDOW} Tháng Rolling Mean"

        # --- Vẽ đồ thị ---
        plt.figure(figsize=(14, 6))
        
        # Vẽ dữ liệu trung bình (để thấy sự dao động/nhiễu)
        plt.plot(resampled_mean.index, resampled_mean.values, 
                 label=f'Trung bình {var.title()} theo {FREQ}', 
                 color='lightblue', alpha=0.6, linestyle='--')
        
        # Vẽ Rolling Mean (đường xu hướng chính)
        plt.plot(rolling_mean.index, rolling_mean.values, 
                 label=f'Đường Xu hướng ({freq_text})', 
                 color='darkred', linewidth=2)
        
        plt.title(f"Câu 2: Xu hướng {var.replace('_', ' ').title()} ({freq_text})")
        plt.xlabel("Thời Gian")
        plt.ylabel(var.replace('_', ' ').title())
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    else:
        print(f"Cột '{var}' không phải là dạng số và đã bị bỏ qua.")
```

```python
# Đảm bảo cột tháng đã được trích xuất
df['Month'] = df.index.month 

# Danh sách các biến đã sử dụng
variables = ['temperature', 'relative_humidity', 'precipitation']
```

```python
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

for var in variables:
    if pd.api.types.is_numeric_dtype(df[var]):
        print(f"\n==================== Mẫu hình Mùa: **{var.replace('_', ' ').title()}** ====================")
        
        # 1. Tính giá trị trung bình của biến theo từng tháng (qua các năm)
        seasonal_mean = df.groupby('Month')[var].mean()

        plt.figure(figsize=(10, 6))
        
        # 2. Vẽ biểu đồ Line
        plt.plot(seasonal_mean.index, seasonal_mean.values, 
                 marker='o', linestyle='-', color='darkgreen', linewidth=2)
        
        plt.title(f"Câu 3: Mẫu hình Mùa (Seasonal) của {var.replace('_', ' ').title()} theo Tháng")
        plt.xlabel("Tháng trong Năm")
        plt.ylabel(f"Giá trị Trung bình theo Mùa ({var.replace('_', ' ').title()})")
        
        # Đặt nhãn trục x là số tháng từ 1 đến 12
        plt.xticks(seasonal_mean.index, [str(i) for i in seasonal_mean.index])
        
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    else:
        print(f"Cột '{var}' không phải là dạng số và đã bị bỏ qua.")
```

```python
# Chọn các cột số (numeric columns) để tính toán tương quan.
# Dựa trên dữ liệu bạn đã cung cấp trước đó, ta chọn các cột số quan trọng
# Thêm 'dew_point' và 'apparent_temperature' để có bức tranh đầy đủ hơn
correlation_columns = [
    'temperature', 
    'relative_humidity', 
    'precipitation',
    'dew_point',
    'apparent_temperature',
    'surface_pressure',
    'wind_speed'
]

# Lọc DataFrame chỉ lấy các cột cần thiết (đảm bảo chúng là dạng số)
# Dùng df.select_dtypes(include='number') để lấy tất cả cột số nếu bạn không chắc chắn
df_corr = df[correlation_columns].copy() 

# 1. Tính Ma trận Tương quan
correlation_matrix = df_corr.corr()

print("Ma trận Tương quan:")
print(correlation_matrix)
```

```python
# Đặt style và figure size
sns.set_style("white")
plt.rcParams["figure.figsize"] = (10, 8)

plt.figure(figsize=(10, 8))

# Vẽ Heatmap
# annot=True: Hiển thị giá trị tương quan trên bản đồ nhiệt
# cmap='coolwarm': Sử dụng bảng màu phân cực (từ lạnh/âm đến ấm/dương)
# fmt='.2f': Định dạng số thập phân với 2 chữ số sau dấu phẩy
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    cmap='coolwarm', 
    fmt='.2f', 
    linewidths=.5,
    cbar_kws={'label': 'Hệ số tương quan'}
)

plt.title('Câu 4: Ma trận Tương quan giữa các Biến Khí hậu')
plt.tight_layout()
plt.show()
```
